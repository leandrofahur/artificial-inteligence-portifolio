# **Simple Linear Regression With Scikit**

```python
# import the necessary pkgs:
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

sns.set()
```

```python
#  guarantee repeatability:
np.random.seed(42)

x_train = np.random.uniform(1, 5, 100)
y_train = 2.5 * x_train + 10

noise = np.random.normal(loc=0.0, scale=1, size=y_train.shape)
y_train = y_train + noise
```

```python
# reshape the data to fit the model:
X_train = x_train.reshape(-1, 1)
y_train = y_train
```

```python
# create the linear regression model:
model = LinearRegression()

# train the model using the training data:
model.fit(X_train, y_train)

print("Slope (m):", model.coef_[0])
print("Intercept (b):", model.intercept_)
```

```python
# make predictions on the training set:
y_pred = model.predict(X_train)

# evaluate the model:
mse = mean_squared_error(y_train, y_pred)
r2 = r2_score(y_train, y_pred)

print("Mean Squared Error:", mse)
print("R^2 Score:", r2)

plt.subplots(1, 1, figsize=(18, 10))
plt.scatter(X_train, y_train, color='blue', label='Sample')
plt.plot(X_train, y_pred, color='red', linewidth=1, label='Predicted')
plt.title('Actual vs Predicted Values')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
